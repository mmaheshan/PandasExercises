## Remove a Series (Column) or Row

```python
oo.drop?
```

```python
# axis = 0 -> drop rows
# axis = 1 -> drop columns
oo.drop('Position', axis=1, inplace=True)
oo.head()
```

```python
# Chaining methods
# When chaining methods, avoid using inplace=True
# The reason is that inplace=True returns None, and thus breaks the chain.
oo = (pd.read_csv(filename, skiprows=5)
      .rename(columns=column_mapper) # Rename columns
      .drop('Position', axis=1)) # Drop Position column
oo.head()
```

```python
# Drop rows
oo = (pd.read_csv(filename, skiprows=5)
      .rename(columns=column_mapper) # Rename columns
      .drop([1,2,3], axis=0)) # Drop rows 1, 2 ,3
oo.head()
```

```python
# Drop rows and columns combined
oo = (pd.read_csv(filename, skiprows=5)
      .rename(columns=column_mapper) # Rename columns
      .drop(labels=[1,2,3], axis=0)
      .drop(columns=['City', 'Position'], axis =1)
      ) # Drop rows 1, 2 ,3
    
oo.head()
```

Using **<span style="color:red;">inline=True</span>** is discouraged. This may get depricated in future. Also chaining is not possible with it.
