## Reorder data based on ordered categories
**_pd.categorical_** function can be used for ordering categories.

```python
medal_order = ['Gold', 'Silver', 'Bronze']
pd.Categorical(oo.Medal, categories=medal_order, ordered=True)
## Reorder data based on ordered categories
```

```python
oo.Medal = pd.Categorical(oo.Medal, categories=medal_order, ordered=True)   
oo.head()
```

Alternatively, you can use **_cat.reorder_categories_**

```python
# Reorder data based on ordered categories
oo.Medal = oo.Medal.cat.reorder_categories(medal_order, ordered=True)   
oo.Medal.cat.categories
oo.head()
```

## Defining dtypes for efficiency

```python
# Define categorical type
medal_order = ['Gold', 'Silver', 'Bronze']
medal_type = pd.CategoricalDtype(categories=medal_order, ordered=True)
# You can also use this way: 
# medal_type = pd.api.types.CategoricalDtype(categories=medal_order, ordered=True)
print(pd.api.types.is_categorical_dtype(medal_type))
column_names = ['Year',
 'City',
 'Sport',
 'Discipline',
 'Athlete_Name',
 'NOC',
 'Gender',
 'Event',
 'Event_Gender',
 'Medal',
 'Position']
medal_type  
```

```python
dtype_mapper = {  
    'Year': 'int16',
    'City': 'category',
    'Sport': 'string',
    'Discipline': 'string',
    'Athlete Name': 'string',
    'NOC': 'category',
    'Gender': 'category',
    'Event': 'string',
    'Event Gender': 'category',
    'Medal': medal_type,
    'Position': 'uint8'  # Nullable integer type
}

# This works: here CSV Is loaded and then header is renamed
# oo = (pd.read_csv(filename, skiprows=5, dtype=dtype_mapper, header=0)
#         .rename(columns=column_mapper) # Rename columns
#     )

# Here both dtype and names parameters are used together
# In this case, you need to use index_col=False to avoid first column being used as index
oo = pd.read_csv(filename, skiprows=5, dtype=dtype_mapper, names=column_names, header=0, index_col=False)
oo.head()
```

## Python functions

```python
def preprocess():
    # Define categorical type
    medal_order = ['Gold', 'Silver', 'Bronze']
    medal_type = pd.CategoricalDtype(categories=medal_order, ordered=True)
    # You can also use this way: 
    # medal_type = pd.api.types.CategoricalDtype(categories=medal_order, ordered=True)
    print(pd.api.types.is_categorical_dtype(medal_type))
    column_names = ['Year', 'City', 'Sport', 'Discipline', 'Athlete_Name', 'NOC', 'Gender', 'Event', 'Event_Gender', 'Medal', 'Position']
    dtype_mapper = {  
        'Year': 'int16',
        'City': 'category',
        'Sport': 'string',
        'Discipline': 'string',
        'Athlete Name': 'string',
        'NOC': 'category',
        'Gender': 'category',
        'Event': 'string',
        'Event Gender': 'category',
        'Medal': medal_type,
        'Position': 'uint8'  # Nullable integer type
    }
    oo = pd.read_csv(filename, skiprows=5, dtype=dtype_mapper, names=column_names, header=0, index_col=False)

    return oo


oo = preprocess()
oo.head()
```

## Creating Series and DataFrames

```python
names = ['John', 'Alice', 'Bob', 'Lucy']
scores = [85, 90, 78, 92]
```

```python
# Convert to dataframe using pd.Series
data = pd.DataFrame({'Name' : pd.Series(names), 'Score' : pd.Series(scores) })
data 
```

```python
# Use zip to create dataframe
res = zip(names, scores)
print(list(res))
```

```python
# Convert to dataframe using zip
data = pd.DataFrame(list(zip(names, scores)), columns=['Name', 'Score'])
data   
```
