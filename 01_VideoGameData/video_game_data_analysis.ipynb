## Melting
### Melting means taking column values and converting them into row values. In this example, NA_Sales, EU_Sales, JP_Sales and Other_Sales columns are converted into row values. This is like **_unpivoting_** table data.

```python
sales
```

```python
sales.melt(id_vars=['Platform', 'Genre'], value_vars=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], var_name='Region', value_name='Sales')    
```

```python
unpivoted_sales = sales.melt(id_vars=['Platform', 'Genre'], value_vars=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], var_name='Region', value_name='Sales')
unpivoted_sales.loc[unpivoted_sales['Region'] == 'NA_Sales', "Region"] = "North America"
unpivoted_sales.loc[unpivoted_sales['Region'] == 'Eu_Sales', "Region"] = "Europe"
unpivoted_sales.loc[unpivoted_sales['Region'] == 'JP_Sales', "Region"] = "Japan"
unpivoted_sales.loc[unpivoted_sales['Region'] == 'Other_Sales', "Region"] = "Other"
unpivoted_sales
```
